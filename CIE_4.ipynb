```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df=pd.read_csv('/content/drive/MyDrive/DATA SET /Mall_Customers.csv')

# data preprocessing
print(" isnull",df.isnull().sum())
print(" drop na  ",df.dropna())

x=df[['CustomerID','Annual Income (k$)','Age']]

# elbow method
squared=[]
k_values=range(1,11)

# creating k means clustering
for k in k_values:
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(x)
  squared.append(kmeans.inertia_)

#Visualization  of Clusters
plt.plot(k_values,squared,marker='o')
plt.xlabel(" Clusters")
plt.ylabel(" squared distances")
plt.title("K_means clustering")
plt.show()
```
